```python
#set up path and dependencies
import pandas as pd

schools_test_results_data = "Resources/schools_complete.csv"
students_test_results_data = "Resources/students_complete.csv"

#read data, create pandas data frames
school_test_results = pd.read_csv(schools_test_results_data)
student_test_results = pd.read_csv(students_test_results_data)

#merge both sets of data into one
total_test_results = pd.merge(school_test_results, student_test_results, how = "left", on = ["school_name", "school_name"])
```

```python
#District Summary: total # of schools, students, budget, avg. math and reading scores
schools_total = total_test_results["school_name"].unique())
students_total = total_test_results["Student ID"].count()
budget_total = schools_test_results_data["budget"].sum()
math_score_avg = total_test_results["math_score"].mean()
reading_score_avg = total_test_results["reading_score"].mean()
students_passing_rate = (math_score_avg + reading_score_avg) / 2

#Percentages: passing math score, reading score, both math and reading 
students_passing_math = total_test_results[(total_test_results["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = students_passing_math / float(students_total) * 100
students_passing_reading = total_test_results[(total_test_results["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = students_passing_reading / float(students_total) * 100

#Data frame with results
district_total = pd.DataFrame({"Total Schools": [schools_total], 
                                 "Total Students": [students_total], 
                                 "Total Budget": [budget_total],
                                 "Average Math Score": [math_score_avg], 
                                 "Average Reading Score": [reading_score_avg],
                                 "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                 "% Overall Passing Rate": [students_passing_rate]})

district_total = district_total[["Total Schools", "Total Students", "Total Budget",
                                 "Average Math Score",
                                 "Average Reading Score",
                                 "% Passing Math",
                                 "% Passing Reading",
                                 "% Overall Passing rate"]]

#formatting 
district_total["Total Students"] = district_total["Total Students"].map("{:,}".format)
district_total["Total Budget"] = district_total["Total Budget"].map("${:,.2f}".format)

district_total
```

```python
#School Summary: School name, type, # of students, general and per student budget
schools_type = school_test_results.set_index(["school_name"])["type"]
students_per_school = total_test_results["school_name"].value_counts()
school_budget = total_test_results.groupby(["school_name"]).mean()["budget"]
school_spending = school_budget / students_per_school

#School Summary: Avg, math score, reading score, % of both passing reading and math
math_students_ps = total_test_results.groupby(["school_name"]).mean()["math_score"]
reading_students_ps = total_test_results.groupby(["school_name"]).mean()["reading_score"]

#conditions to determine scores
math_schools_ps = total_test_results[(total_test_results["math_score"] >= 70)]
reading_schools_ps = total_test_results[(total_test_results["reading_score"] >= 70)] 

student_ps_math = math_schools_ps.groupby(["school_name"]).count()["student_name"] / students_per_school * 100
student_ps_reading = reading_schools_ps.groupby(["school_name"].count()["student_name"] / students_per_school * 100
students_passing_rate = (student_ps_math + student_ps_reading) / 2

#Data Frame with results / formatting
school_summary = pd.DataFrame({"School Type": schools_type,
                               "Total Students": students_per_school,
                               "Total School Budget": school_budget,
                               "Per Student Budget": school_spending,
                               "Average Math Score": math_students_ps,
                               "Average Reading Score": reading_students_ps,
                               "% Passing Math": math_schools_ps,
                               "%Passing Reading": reading_schools_ps,
                               "%Overall Passing Rate": passing_rate})

school_summary = school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                 "Average Math Score", "Average Reading Score",
                                 "% Passing Math", "% Passing Reading",
                                 "% Overall Passing Rate"]]

school_summary["Total School Budget"] = school_summary["Total SChool Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("{:,.2f}".format)

school_summary
```
